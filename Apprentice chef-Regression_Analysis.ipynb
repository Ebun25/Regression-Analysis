DATA ANALYSIS
To gain insight into the factors affecting revenue of the company, a thorough analysis was conducted to determine the correlation between various variables and revenue. The results of this analysis revealed that some variables were positively correlated with revenue, while others were negatively correlated or not correlated at all. 
To gain further insights into the data, a histogram was created to examine the skewness of the data. The histogram revealed that the data was skewed. To correct this issue, the regression model was used to transform certain variables into logarithmic form, which was then saved in the dataset. A new histogram was then created, which showed that the data was now well-skewed.
The ordinary Least Square (OLS) model was used, utilizing the statsmodel library, to analyze the relationship between the logarithmic transformations of revenue and the various continuous variables in the dataset. The logarithmic variables studied include average time on site, total meals ordered, total photo viewed, contact with customer service, average mean rating, and unique meals purchased. 
The result of the analysis indicate that the model has a moderate level of explanatory power, as indicated by the R-squared and adjusted R-squared values of 0.449 and 0.448, respectively. Additionally, the p-value of 0.000 suggests that the factors included in the model are statistically significant in explaining the logarithmic form of the companyâ€™s revenue. These results suggest that the independent variables have a significant impact on revenue.

The "train_test_split" function from scikit-learn is used to split the data into two parts, a training set and a testing set. The "train_test_split" function splits the data into training and testing sets, with the test set size being specified by the argument "test_size" and set to 0.25. The argument "random_state" is set to 219, which ensures that the data is split in a consistent manner each time the code is run. The purpose of splitting the data into a training set and a testing set is to evaluate the performance of the model

Fitting the model to the training data (x_train and y_train) using the fit method of the Linear Regression object. The model is then used to make predictions on the test data (x_test) using the predict method.

The performance of the model is evaluated by calculating the R-squared values for both the training and test datasets. The R-squared value is a measure of how well the model fits the data. It ranges from 0 to 1, with higher values indicating a better fit. In this case, the training score is 0.6681 and the testing score is 0.673.

The train-test gap is calculated as the difference between the training and testing scores. In this case, the gap is 0.0049, which is relatively small. This suggests that the model has good generalization performance, meaning it can make accurate predictions on unseen data.

The code then lists the coefficients for each independent variable in the model, along with their corresponding feature names. These coefficients represent the strength and direction of the relationship between the independent variable and the dependent variable. Positive coefficients indicate a positive relationship (as the independent variable increases, so does the dependent variable), while negative coefficients indicate a negative relationship (as the independent variable increases, the dependent variable decreases). The intercept is the constant value in the model, which represents the expected value of the dependent variable when all independent variables are equal to zero.
From the analysis  of the linear regression analysis, where each feature (i.e. independent variable) in the model is paired with its corresponding coefficient. The coefficients represent the relationship between the feature and the dependent variable.The intercept (i.e. the constant term) is 4.07. This means that when all the features are zero, the estimated value of the dependent variable is 4.07.The coefficient for "log arithmetic average time per site visit" is 0.48, which means that for a one unit increase in the " log arithmetic average time per site visit " feature, the estimated value of the dependent variable increases by 0.48. Similarly, the coefficients for the other features represent the relationship between each feature and the dependent variable.

The ARD regression which was also trained on some data was used in measuring the accuracy of the model's predictions by comparing them to the actual values. The accuracy of the model is being measured using the R-square metric.
The training score of the model is 0.6681, which means that the model's predictions are off by an average of 0.6681 when compared to the actual values in the training data. The testing score of the model is 0.6726, which means that the model's predictions are off by an average of 0.6726 when compared to the actual values in the testing data. The train-test gap is 0.0045, which means that there is a small difference between the training score and the testing score.
A Decision Tree Regressor was trained using a randomized search cross-validation approach. The random search cross-validation approach randomly samples a set of hyperparameters from the provided parameter grid and trains the model with those hyperparameters. The performance of the model is then evaluated using cross-validation with 5 folds. This process is repeated 1000 times and the hyperparameters that result in the best performance (highest AUC score) are saved as the "best parameters."

The final output indicates that the best training AUC achieved was 0.6558, which is the average cross-validated AUC score across all 5 folds for the best hyperparameters. The AUC of 0.6558 means the model's performance is better than random.

This code trains a Decision Tree Regressor using the hyperparameters that were found to be optimal in the previous randomized search cross-validation step. The model is fit to the training data using the fit method. The predict method is then used to make predictions on the test data.

The code then calculates and outputs the model's score on both the training data and the test data using the score method. The score method returns the coefficient of determination (R^2) which is a measure of how well the model fits the data. A value of 1.0 indicates a perfect fit, and a value of 0.0 indicates a constant model that only predicts the mean of the target values. So, a score of 0.7584 on the training data and 0.7123 on the test data indicates that the model has a good fit to the training data.
Finally, the code calculates the difference between the training score and the test score, which is 0.0460 in this case. This difference, also known as generalization error, measures how well the model generalizes to new unseen data.

```python
# importing libraries
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # enhanced data visualization
import statsmodels.formula.api as smf # regression modeling
from sklearn.model_selection import train_test_split # train/test split
import sklearn.linear_model # linear modeling in scikit-learn
```

```python
# x-variables for normal y
x_var = ['NAME','EMAIL','FIRST_NAME','FAMILY_NAME','TOTAL_MEALS_ORDERED','UNIQUE_MEALS_PURCH','CONTACTS_W_CUSTOMER_SERVICE',
'PRODUCT_CATEGORIES_VIEWED','AVG_TIME_PER_SITE_VISIT','CANCELLATIONS_AFTER_NOON','PC_LOGINS','MOBILE_LOGINS','WEEKLY_PLAN','LATE_DELIVERIES ',               
'AVG_PREP_VID_TIME','LARGEST_ORDER_SIZE','AVG_MEAN_RATING','TOTAL_PHOTOS_VIEWED']

# x-variables for log y
x_var_log_y = ['log_AVG_TIME_PER_SITE_VISIT','log_TOTAL_MEALS_ORDERED',
                   'log_TOTAL_PHOTOS_VIEWED','log_CONTACTS_W_CUSTOMER_SERVICE',
               'log_AVG_MEAN_RATING','log_UNIQUE_MEALS_PURCH']
```

```python
# preparing for scikit-learn

# preparing x-variables from the OLS model
x_data = chef.loc[ : ,x_var_log_y]



# preparing response variable
y_data = chef.loc[ : , 'log_REVENUE']


###############################################
## setting up more than one train-test split ##
###############################################
x_train, x_test, y_train, y_test = train_test_split(
            x_data, # x-variables (can change this)
            y_data, # y-variable  (can change this)
            test_size    = 0.25,
            random_state = 219)
```

```python
# Setting a model name
model_name = "Linear Regression"

# Creating a model object
model = sklearn.linear_model.LinearRegression()

# training the  data for fittings
model_fit = model.fit(x_train, y_train)

# PREDICTING the new data
model_pred = model.predict(x_test)

# SCORING the results
model_train_score = model.score(x_train, y_train).round(4) # using R-square
model_test_score  = model.score(x_test, y_test).round(4)   # using R-square
model_gap         = abs(model_train_score - model_test_score).round(4)

# Displaying results
print('Training Score:', model_train_score)
print('Testing Score :', model_test_score)
print('Train-Test Gap:', model_gap)
```

```python
# zipping each of the feature name to the coefficient
model_coefficients = zip(x_train.columns,
                         model.coef_.round(decimals = 2))

coefficient_lst = [('intercept', model.intercept_.round(decimals = 2))]

for coefficient in model_coefficients:
    coefficient_lst.append(coefficient)
    
for pair in coefficient_lst:
    print(pair)
```

```python
# dynamically printing results
ols_model =  f"""\
Model Name:     {model_name}
Train_Score:    {model_train_score}
Test_Score:     {model_test_score}
Train-Test Gap: {model_gap}

Coefficents
-----------
{pd.DataFrame(data = coefficient_lst, columns = ["Feature", "Coefficient"])}"""

print(ols_model)
```

```python
# Setting, fitting and predicting a model name
model_name = "ARD Regression"
model = sklearn.linear_model.ARDRegression()
model_fit = model.fit(x_train, y_train)
model_pred = model.predict(x_test)


# SCORING the results
model_train_score = model.score(x_train, y_train).round(4) # using R-square
model_test_score  = model.score(x_test, y_test).round(4)   # using R-square
model_gap         = abs(model_train_score - model_test_score).round(4)


# displaying results
print('Training Score :', model_train_score)
print('Testing Score  :', model_test_score)
print('Train-Test Gap :', model_gap)
```
